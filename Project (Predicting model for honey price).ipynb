```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import statsmodels.formula.api as smf
```

```python
from sklearn.linear_model import LinearRegression
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import adam
```

```python
x = data.drop(["Price"] , axis = 1)
y = data["Price"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.20 , random_state = 42)
```

```python
x_train.shape , x_test.shape
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
model = LinearRegression()
```

```python
model.fit(x_train , y_train)
```

```python
y_model_pred = model.predict(x_test) 
```

```python
y_model_pred
```

```python
mse = np.mean((y_test-y_model_pred)**2)
```

```python
rmse = np.sqrt(mse)
```

```python
print("rmse_linear_model:",rmse)
```

```python
pd.DataFrame({"y_test":y_test , "y_model_pred":y_model_pred})
```

```python
model_tree = DecisionTreeRegressor(criterion = "friedman_mse" , max_depth = 5)
```

```python
model_tree.fit(x_train , y_train)
```

```python
y_tree_pred = model_tree.predict(x_test)
```

```python
y_tree_pred
```

```python
mse_tree = np.mean((y_test-y_tree_pred)**2)
```

```python
rmse_tree = np.sqrt(mse_tree)
print("rmse_tree_model:",rmse_tree)
```

```python
model_neural = Sequential()
model_neural.add(Dense(11 , input_dim = 10 , activation = "relu")) ## input layer
model_neural.add(Dense(12 , activation = "relu" )) ##1st hidden layer
model_neural.add(Dense(24  , activation = "relu")) ## 2nd hidden layer
model_neural.add(Dense(36 , activation = "relu")) ## 3rd hidden layer
model_neural.add(Dense(18 ,activation = "relu")) ## 4th hidden layer
model_neural.add(Dense(9 , activation = "relu")) ## 5 th hidden layer
model_neural.add(Dense(1 , activation = "linear")) ## output layer

                 
```

```python
model_neural.compile(loss = "mean_squared_error" , optimizer = "adam")
```

```python
history = model_neural.fit(x_train , y_train , batch_size = 10 , epochs = 25 , validation_split = 0.20)
```

```python
from matplotlib import pyplot
```

```python
pyplot.plot(history.history["loss"] , label = "train")
pyplot.plot(history.history["val_loss"],label = "test")
plt.title("mean_squared_error")
```
